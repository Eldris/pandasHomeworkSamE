```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
premstats_df = pd.read_csv('premStats.csv', sep=',')
```

```python
print(premstats_df)
```

```python
premstats_df.keys()
```

```python
#Creates a variable containing the top 10 goalscorers
topScorers = premstats_df.sort_values(by=['Goals'], ascending=False)[0:10]
```

```python
#Creates a bar chart with the breakdown of top players total goals
topScorers[['Name', 'Goals', 'Goals with right foot', 'Goals with left foot', 'Headed goals']].plot.bar(x='Name')
```

```python
#Stacks the data and I also removed Goals as it showed the data incorrectly
topScorers[['Name', 'Goals with right foot', 'Goals with left foot', 'Headed goals']].plot.bar(x='Name', stacked = True)
```

```python
premstats_df['Nationality'].value_counts().plot.pie(x='Nationality')
```

```python
#Reduced the amount of nationalities for clarity on the Pie Chart
topNationalities = premstats_df['Nationality'].value_counts()[0:20]
print(topNationalities)
```

```python
topNationalities.plot.pie(x='Nationality')
```

```python
#Shows only players from Leeds
premstats_df[(premstats_df['Club'] == 'Leeds-United')]
```

```python
#Displays the assist to goal ratio of the top 10
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

for key in topScorers['Name'].unique():
    print(key)

    axes.scatter(x=topScorers[topScorers['Name'] == key]['Goals'],
                 y=topScorers[topScorers['Name'] == key]['Assists'],
                s=15,
                label=key)
    
    plt.legend()
```

```python
#Created a smaller pool as the data was too big for a Goals/Assists Scatter Graph
premstats_df2 = premstats_df[(premstats_df['Goals'] > 10)&(premstats_df['Assists'] > 10)]
```

```python
#Compares all players who registered over 10 Goals and Assists
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 10))

for key in premstats_df2['Name'].unique():
    print(key)

    axes.scatter(x=premstats_df2[premstats_df2['Name'] == key]['Goals'],
                 y=premstats_df2[premstats_df2['Name'] == key]['Assists'],
                s=15,
                label=key)
    
    plt.legend()
```
